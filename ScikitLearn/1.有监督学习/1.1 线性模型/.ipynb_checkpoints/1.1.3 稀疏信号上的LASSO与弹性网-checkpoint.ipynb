估计Lasso和弹性网络回归对手动产生的、被破坏的、加了噪声的稀疏信号进行建模。估计系数与真实值作了比较。

```python
print(__doc__)

import numpy as np
import matplotlib.pyplot as plt
# 设置中文字体支持
plt.rcParams["font.family"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False  # 正确显示负号

from sklearn.metrics import r2_score
```

```python
# #############################################################################
# Generate some sparse data to play with
np.random.seed(42)

n_samples, n_features = 50, 100
X = np.random.randn(n_samples, n_features)
```

```python
# Decreasing coef w. alternated signs for visualization
idx = np.arange(n_features)
coef = (-1) ** idx * np.exp(-idx / 10)
coef[10:] = 0  # sparsify coef
y = np.dot(X, coef)
```

```python
# Add noise
y += 0.01 * np.random.normal(size=n_samples)
```

```python
# Split data in train set and test set
n_samples = X.shape[0]
X_train, y_train = X[:n_samples // 2], y[:n_samples // 2]
X_test, y_test = X[n_samples // 2:], y[n_samples // 2:]
```

```python
# #############################################################################
# Lasso
from sklearn.linear_model import Lasso

alpha = 0.1
lasso = Lasso(alpha=alpha)

y_pred_lasso = lasso.fit(X_train, y_train).predict(X_test)
r2_score_lasso = r2_score(y_test, y_pred_lasso)
print(lasso)
print("r^2 on test data : %f" % r2_score_lasso)
```

```python
# #############################################################################
# ElasticNet
from sklearn.linear_model import ElasticNet

enet = ElasticNet(alpha=alpha, l1_ratio=0.7)

y_pred_enet = enet.fit(X_train, y_train).predict(X_test)
r2_score_enet = r2_score(y_test, y_pred_enet)
print(enet)
print("r^2 on test data : %f" % r2_score_enet)
```

```python
# 绘制系数对比图
plt.figure(figsize=(12, 7))

# Elastic Net 系数
m_enet, s_enet, _ = plt.stem(
    np.where(enet.coef_)[0],
    enet.coef_[enet.coef_ != 0],
    markerfmt='x',
    linefmt='#2ca02c',  # 直接设置线条颜色
    basefmt='None',     # 不显示基线
    label='弹性网络系数'
)
plt.setp(m_enet, color='#2ca02c', markersize=8)  # 设置标记颜色和大小

# Lasso 系数
m_lasso, s_lasso, _ = plt.stem(
    np.where(lasso.coef_)[0],
    lasso.coef_[lasso.coef_ != 0],
    markerfmt='x',
    linefmt='#ff7f0e',  # 直接设置线条颜色
    basefmt='None',     # 不显示基线
    label='Lasso系数'
)
plt.setp(m_lasso, color='#ff7f0e', markersize=8)  # 设置标记颜色和大小

# 真实系数
m_true, s_true, _ = plt.stem(
    np.where(coef)[0],
    coef[coef != 0],
    markerfmt='o',
    linefmt='#1f77b4',  # 直接设置线条颜色
    basefmt='None',     # 不显示基线
    label='真实系数'
)
plt.setp(m_true, color='#1f77b4', markerfacecolor='none', markersize=8)  # 空心圆标记

# 添加网格和标签
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xlabel('特征索引', fontsize=12)
plt.ylabel('系数值', fontsize=12)
plt.title(f"Lasso R²: {r2_score_lasso:.3f}, 弹性网络 R²: {r2_score_enet:.3f}", fontsize=14)
plt.legend(loc='best', fontsize=10)
plt.tight_layout()

plt.show()
```
