在支持向量机（如 SVC、NuSVC）中，decision_function_shape 参数用于指定多类别分类时的决策函数形式，其取值 'ovr'（One-Vs-Rest）和 'ovo'（One-Vs-One）是两种经典的多类别分类策略。以下从原理、计算过程、优缺点等方面详细解释：

## 一、核心背景：SVM 与多类别分类

SVM 本质是二分类模型（仅能区分两个类别）。当面对 K 个类别（K≥3）时，需要通过策略将其转化为多个二分类问题的组合，'ovr' 和 'ovo' 就是两种最常用的策略。

## 二、'ovr'（One-Vs-Rest，一对多）

<B>原理<B>

对 K 个类别，训练 K 个二分类器：
* 第 i 个分类器：将 “类别 i” 视为正类，将 “所有其他类别（1,2,...,i-1,i+1,...,K）” 视为负类。
* 预测时，输入样本会被 K 个分类器分别打分（通过 decision_function 输出的距离值），最终选择打分最高的类别作为结果。

<B>示例（3个类别：A、B、C）<B>

* 训练 3 个分类器：
1. 分类器 1：A（正） vs B+C（负）
2. 分类器 2：B（正） vs A+C（负）
3. 分类器 3：C（正） vs A+B（负）
* 预测某样本时，若分类器 1 打分 0.8，分类器 2 打分 -0.3，分类器 3 打分 0.5，则最终预测为 A（最高分为 0.8）。

## 三、'ovo'（One-Vs-One，一对一）

<B>原理<B>

对 K 个类别，训练 K(K-1)/2 个二分类器*：
* 每个分类器仅针对一对类别（如类别 i 和类别 j，i<j），专门学习区分这两个类别。
* 预测时，输入样本会被所有分类器投票：每个分类器判定样本属于其对应的两个类别中的一个，最终得票最多的类别作为结果。

<B>示例（3 个类别：A、B、C）<B>

* 训练 3 个分类器（3=3*2/2）：
  * 分类器 1：A vs B
  * 分类器 2：A vs C
  * 分类器 3：B vs C
* 预测某样本时：
  * 若分类器 1 判为 A，分类器 2 判为 A，分类器 3 判为 B，则 A 得 2 票，B 得 1 票，最终预测为 A。

## 四、'ovr' 与 'ovo' 的对比

|维度	        |'ovr'（一对多）	|'ovo'（一对一）|
|   :----        |    :----           |    :----       |
|分类器数量	    |K 个（K 为类别数）	|K*(K-1)/2 个（远多于 ovr）|
|训练时间	    |较短（分类器少）	|较长（分类器多，尤其 K 较大时）|
|预测时间	    |较快（只需 K 次计算）	|较慢（需 K*(K-1)/2 次计算）|
|决策函数输出	|每个样本输出 K 个值（对应每个类别）	|每个样本输出 1 个值（仅用于二分类时；多类别时需额外处理投票）|
|类别不平衡适应性	|较差（负类样本量远大于正类）	|较好（每对类别样本量较均衡）|
|内存占用	    |较小（存储 K 个模型）	|较大（存储更多模型）|
|适用场景	    |类别数较多（如 K=100）	|类别数较少（如 K=5）|

## 五、实际应用建议

1. 默认选择：scikit-learn 中 decision_function_shape 的默认值为 'ovr'，因其训练和预测效率更高，尤其适合类别数较多的场景。
2. 小类别数（K≤5）：可尝试 'ovo'，其分类精度可能略高（因每对类别单独优化）。
3. 类别不平衡数据：优先 'ovo'，避免 'ovr' 中负类样本过多导致的分类器偏向。
4. 决策函数使用：若需使用 decision_function 输出的分数（如计算置信度），'ovr' 更直观（直接对应每个类别的得分），而 'ovo' 需通过投票机制转换结果。

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import time

# 加载鸢尾花数据集（4个特征，3个类别）
iris = datasets.load_iris()
X = iris.data[:, :2]  # 仅取前两个特征便于可视化
y = iris.target

# 数据划分与标准化
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 创建模型（ovr 和 ovo）
models = {
    'SVC (ovr)': SVC(decision_function_shape='ovr', kernel='rbf', gamma='scale'),
    'SVC (ovo)': SVC(decision_function_shape='ovo', kernel='rbf', gamma='scale')
}

results = {}

# 创建画布
plt.figure(figsize=(15, 6))

# 网格用于绘制决策边界
h = .02
x_min, x_max = X_train_scaled[:, 0].min() - 1, X_train_scaled[:, 0].max() + 1
y_min, y_max = X_train_scaled[:, 1].min() - 1, X_train_scaled[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

for i, (name, model) in enumerate(models.items()):
    # 训练模型并记录时间
    start_time = time.time()
    model.fit(X_train_scaled, y_train)
    train_time = time.time() - start_time
    
    # 预测并记录时间
    start_time = time.time()
    y_pred = model.predict(X_test_scaled)
    pred_time = time.time() - start_time
    
    # 计算准确率
    accuracy = accuracy_score(y_test, y_pred)
    
    # 记录结果
    results[name] = {
        'accuracy': accuracy,
        'train_time': train_time,
        'pred_time': pred_time,
        'n_support_vectors': model.n_support_,  # 每类的支持向量数量
        'n_classifiers': model.decision_function_shape  # 分类器数量
    }
    
    # 绘制决策边界
    plt.subplot(1, 2, i + 1)
    Z = model.decision_function(np.c_[xx.ravel(), yy.ravel()])
    
    # 对于ovr，decision_function输出形状为(n_samples, n_classes)
    # 对于ovo，decision_function输出形状为(n_samples, n_classes*(n_classes-1)/2)
    if name.endswith('(ovr)'):
        # 直接取最大值对应的类别
        Z = np.argmax(Z, axis=1)
    else:
        # 对于ovo，需要通过投票机制转换
        Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    
    Z = Z.reshape(xx.shape)
    
    # 绘制等高线和数据点
    plt.contourf(xx, yy, Z, alpha=0.8, levels=np.unique(y).size - 1, cmap=plt.cm.Paired)
    plt.contour(xx, yy, Z, levels=np.unique(y), linewidths=2, colors='k')
    
    # 绘制训练数据点
    plt.scatter(X_train_scaled[:, 0], X_train_scaled[:, 1], c=y_train, 
                edgecolors='k', s=50, cmap=plt.cm.Paired)
    
    plt.title(f'{name}\nAccuracy: {accuracy:.4f}, Train Time: {train_time:.4f}s')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')

plt.tight_layout()
plt.show()

# 打印详细结果对比
print("\n===== 模型性能对比 =====")
print("{:<15} {:<10} {:<12} {:<12} {:<15}".format(
    "模型", "准确率", "训练时间(s)", "预测时间(s)", "支持向量数量"))
for name, result in results.items():
    print("{:<15} {:<10.4f} {:<12.6f} {:<12.6f} {}".format(
        name, result['accuracy'], result['train_time'], result['pred_time'], 
        result['n_support_vectors']))

# 打印决策函数形状和分类器数量
print("\n===== 决策函数与分类器 =====")
print(f"SVC (ovr) 决策函数输出形状: {models['SVC (ovr)'].decision_function(X_test_scaled[:10]).shape}")
print(f"SVC (ovo) 决策函数输出形状: {models['SVC (ovo)'].decision_function(X_test_scaled[:10]).shape}")
print(f"SVC (ovr) 分类器数量: {len(models['SVC (ovr)'].classes_)}")
print(f"SVC (ovo) 分类器数量: {len(models['SVC (ovo)'].classes_) * (len(models['SVC (ovo)'].classes_) - 1) // 2}")    
```
