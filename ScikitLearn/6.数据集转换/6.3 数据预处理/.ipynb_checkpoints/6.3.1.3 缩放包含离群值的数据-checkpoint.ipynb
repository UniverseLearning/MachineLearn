RobustScaler 是 scikit-learn 中专门用于处理包含异常值（Outliers）数据的特征缩放工具。与其他缩放器（如 StandardScaler、MinMaxScaler）不同，它基于中位数和四分位数进行缩放，对异常值具有更强的鲁棒性（Robustness）。下面从原理、使用方法、适用场景和优缺点等方面进行详细解释：

```python
import numpy as np
from sklearn.preprocessing import RobustScaler, StandardScaler
import matplotlib.pyplot as plt

# 创建包含异常值的数据
X = np.array([[1], [2], [3], [4], [5], [100]])  # 最后一个值是异常值

# 使用 RobustScaler 缩放
scaler_robust = RobustScaler()
X_robust = scaler_robust.fit_transform(X)

# 使用 StandardScaler 缩放（作为对比）
scaler_standard = StandardScaler()
X_standard = scaler_standard.fit_transform(X)

# 打印结果
print("原始数据:", X.flatten())
print("RobustScaler 缩放后:", X_robust.flatten())
print("StandardScaler 缩放后:", X_standard.flatten())

# 可视化对比
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.title('原始数据')
plt.boxplot(X)
plt.subplot(1, 3, 2)
plt.title('RobustScaler 缩放后')
plt.boxplot(X_robust)
plt.subplot(1, 3, 3)
plt.title('StandardScaler 缩放后')
plt.boxplot(X_standard)
plt.tight_layout()
plt.show()
```
