在高斯变换（将数据转换为接近高斯分布的过程）中，Yeo-Johnson 变换和 Box-Cox 变换是两种常用的幂变换方法。它们通过引入参数化的数学公式，调整数据分布的形状，以减少偏度（分布不对称性）并稳定方差，最终使数据更接近高斯分布（正态分布）。

## 一、Yeo-Johnson 变换与 Box-Cox 变换的核心原理

    两种变换均通过寻找最优参数 λ（lambda）来实现分布矫正，λ 的取值决定了变换的强度和方向。不同的是，它们的适用数据范围和具体公式不同。

## 二、代码示例：对比两种变换的效果
    下面通过具体代码展示如何使用 PowerTransformer 实现两种变换，并可视化变换前后的分布差异。
    

### 步骤 1：准备数据
    生成三种典型的非高斯分布数据（右偏分布、左偏分布、含负数的分布），用于测试两种变换的效果。

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PowerTransformer
from scipy import stats
```

```python
# 设置随机种子，保证结果可复现
np.random.seed(42)
```

```python
# 1. 右偏分布（如收入数据，适合Box-Cox和Yeo-Johnson）
# 生成对数正态分布（指数化正态分布，天然右偏）
right_skewed = np.random.lognormal(mean=1, sigma=0.8, size=1000)  # 全为正数
```

```python
# 2. 左偏分布（如考试分数集中在高分段，适合Yeo-Johnson）
# 生成负的对数正态分布（左偏）
left_skewed = -np.random.lognormal(mean=1, sigma=0.8, size=1000)  # 包含负数
```

```python
# 3. 含0的数据（仅Yeo-Johnson适用）
mixed_data = np.concatenate([
    np.random.lognormal(mean=1, sigma=0.5, size=800),  # 正数
    np.zeros(100),  # 0
    -np.random.lognormal(mean=0.5, sigma=0.3, size=100)  # 负数
])
```

## 步骤 2：应用变换并可视化
    使用 PowerTransformer 分别对数据进行 Box-Cox 和 Yeo-Johnson 变换，对比变换前后的分布。

```python
# 设置中文字体
plt.rcParams["font.family"] = ["SimHei"]
# 定义可视化函数：对比原始数据和变换后的数据分布
def plot_transform(data, transformed_boxcox=None, transformed_yeojohnson=None, title=""):
    plt.figure(figsize=(12, 4))
    
    # 原始数据分布
    plt.subplot(1, 3, 1)
    plt.hist(data, bins=30, edgecolor='k')
    plt.title(f"原始数据\n偏度: {stats.skew(data):.2f}")
    plt.xlabel("值")
    plt.ylabel("频数")
    
    # Box-Cox变换后分布（若适用）
    if transformed_boxcox is not None:
        plt.subplot(1, 3, 2)
        plt.hist(transformed_boxcox, bins=30, edgecolor='k')
        plt.title(f"Box-Cox变换后\n偏度: {stats.skew(transformed_boxcox):.2f}")
        plt.xlabel("值")
    
    # Yeo-Johnson变换后分布
    plt.subplot(1, 3, 3)
    plt.hist(transformed_yeojohnson, bins=30, edgecolor='k')
    plt.title(f"Yeo-Johnson变换后\n偏度: {stats.skew(transformed_yeojohnson):.2f}")
    plt.xlabel("值")
    
    plt.suptitle(title, fontsize=14)
    plt.tight_layout(rect=[0, 0, 1, 0.95])  # 调整标题位置
    plt.show()
```

```python
# 1. 测试右偏分布（仅含正数，两种变换均适用）
pt_boxcox = PowerTransformer(method='box-cox', standardize=False)
pt_yeojohnson = PowerTransformer(method='yeo-johnson', standardize=False)

# Box-Cox仅适用于正数，可直接转换
right_boxcox = pt_boxcox.fit_transform(right_skewed.reshape(-1, 1)).flatten()
# Yeo-Johnson也适用
right_yeojohnson = pt_yeojohnson.fit_transform(right_skewed.reshape(-1, 1)).flatten()

plot_transform(
    right_skewed, 
    transformed_boxcox=right_boxcox, 
    transformed_yeojohnson=right_yeojohnson,
    title="右偏分布（仅正数）的变换效果"
)
```

```python
# 2. 测试左偏分布（含负数，仅Yeo-Johnson适用）
try:
    # 尝试用Box-Cox转换含负数的数据（会报错）
    left_boxcox = pt_boxcox.fit_transform(left_skewed.reshape(-1, 1)).flatten()
except ValueError as e:
    print(f"Box-Cox转换左偏分布时的错误：{e}")

# Yeo-Johnson可处理负数
left_yeojohnson = pt_yeojohnson.fit_transform(left_skewed.reshape(-1, 1)).flatten()

plot_transform(
    left_skewed, 
    transformed_boxcox=None,  # Box-Cox不适用
    transformed_yeojohnson=left_yeojohnson,
    title="左偏分布（含负数）的变换效果"
)
```

```python
# 3. 测试含0的数据（仅Yeo-Johnson适用）
try:
    # 尝试用Box-Cox转换含0的数据（会报错）
    mixed_boxcox = pt_boxcox.fit_transform(mixed_data.reshape(-1, 1)).flatten()
except ValueError as e:
    print(f"Box-Cox转换含0的数据时的错误：{e}")

# Yeo-Johnson可处理0和负数
mixed_yeojohnson = pt_yeojohnson.fit_transform(mixed_data.reshape(-1, 1)).flatten()

plot_transform(
    mixed_data, 
    transformed_boxcox=None,  # Box-Cox不适用
    transformed_yeojohnson=mixed_yeojohnson,
    title="含0和负数的数据的变换效果"
)
```

## 三、代码说明与结果分析

1. 参数 standardize：PowerTransformer 默认会对变换后的数据进行标准化（均值为 0，方差为 1），设置 standardize=False 可保留原始变换后的尺度，便于观察分布形状变化。
2. 偏度（Skew）：分布不对称性的指标（高斯分布的偏度为 0）。变换后偏度越接近 0，说明分布越接近高斯分布。
3. 适用场景验证：
    *  右偏分布（仅正数）：两种变换均可有效降低偏度，但 Box-Cox 更简洁；
    *  左偏分布（含负数）：Box-Cox 因数据含负数而报错，Yeo-Johnson 可成功矫正；
    *  含 0 的数据：Box-Cox 因数据含 0 而报错，Yeo-Johnson 可正常处理。

## 四、总结


| 特性	| Box-Cox 变换	 |Yeo-Johnson 变换|
| ---- | ---- | ---- |
|适用数据范围	|仅正数（x > 0）	|所有实数（x ∈ R）|
|公式复杂度	|简单	|复杂（需处理负数和 0）|
|对偏度的矫正能力	|适用于右偏分布	|适用于左偏、右偏及混合分布|
|易用性	|仅当数据全为正时使用	|通用场景优先选择|

#### 实际应用建议：
* 若数据确认全为正数，可尝试 Box-Cox（可能更高效）；
* 若数据含 0、负数，或不确定数据范围，优先使用 Yeo-Johnson；
* 变换后需通过可视化（直方图、Q-Q 图）或偏度值验证效果，避免盲目依赖。
